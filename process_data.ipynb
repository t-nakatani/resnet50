```python
fnames_base_tr = list(spt[spt['tr/ts'] == 'tr']['fname'])
fnames_base_ts = list(spt[spt['tr/ts'] == 'ts']['fname'])

fnames_novel_tr = list(qry[qry['tr/ts'] == 'tr']['fname'])
fnames_novel_ts = list(qry[qry['tr/ts'] == 'ts']['fname'])


labels_base_tr = list(spt[spt['tr/ts'] == 'tr']['label'])
labels_base_ts = list(spt[spt['tr/ts'] == 'ts']['label'])

labels_novel_tr = list(qry[qry['tr/ts'] == 'tr']['label'])
labels_novel_ts = list(qry[qry['tr/ts'] == 'ts']['label'])
```

```python
dic_base_tr = dict([(fname, label) for fname, label in zip(fnames_base_tr, labels_base_tr)])
dic_base_ts = dict([(fname, label) for fname, label in zip(fnames_base_ts, labels_base_ts)])

dic_novel_tr = dict([(fname, label) for fname, label in zip(fnames_novel_tr, labels_novel_tr)])
dic_novel_ts = dict([(fname, label) for fname, label in zip(fnames_novel_ts, labels_novel_ts)])
```
